```python
from Model import Feature_ViT, MultiView_Classifier
import torch
from ImgDataset import MultiviewImgDataset
```

```python
train_set = MultiviewImgDataset("data/ModelNet40-12-split/train", num_views=12)
train_loader = torch.utils.data.DataLoader(train_set, batch_size=1, shuffle=True)
```

```python
feature_vit = Feature_ViT(num_views=12)
multi_view_model = MultiView_Classifier(num_views=12, num_classes=40)
```

```python
views = next(iter(train_loader))[1]
B, V, C, H, W = views.shape
views_flat = views.view(B * V, C, H, W)

feature_vit = Feature_ViT(num_views=V)
features = feature_vit(views_flat)
```

```python
next(iter(train_loader))
```

```python
cls_tokens = features[:, 0, :]
cls_tokens = cls_tokens.view(B, V, -1)
```

```python
logits = multi_view_model(cls_tokens) 
logits.shape
```

```python
torch.max(logits, dim=1)
#get index of max value
_, preds = torch.max(logits, dim=1)
preds
```
